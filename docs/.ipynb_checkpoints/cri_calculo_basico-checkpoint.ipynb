# Cálculo Básico del Confidence and Risk Index (CRI)

Este notebook muestra cómo calcular el CRI para una serie temporal simple.
Incluye:

- Generación de una serie de precios simulada
- Cálculo paso a paso del CRI
- Visualización del precio y del CRI
- Interpretación del resultado

Este ejemplo sirve como base para comprender el funcionamiento del indicador antes de aplicarlo a activos reales.




```python
# Importación de librerias.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
# Instalar yfinance si es necesario.
# !pip install yfinance

# Generación de una serie de precios simulada
np.random.seed(42)

# Simulación de un mercado alcista suave
ticker = "^GSPC"  # S&P 500
data = yf.download(ticker, start="2015-01-01", progress=False, auto_adjust=True)
df = data[["Close"]].rename(columns={"Close": "price"})
df.head()


# Funciones auxiliares para el CRI

def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

# Cálculo del CRI (versión simple)
# Suavizado
df["smooth"] = smooth(df["price"])

# Volatilidad local
df["vol"] = local_volatility(df["price"])

# Coherencia direccional
df["dir"] = directional_consistency(df["price"])

# Normalizaciones simples
vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
dir_norm = df["dir"].rolling(10, min_periods=1).mean()

# CRI conceptual
df["CRI"] = (0.5 * (1 - vol_norm)) + (0.5 * dir_norm)
df["CRI"] = df["CRI"].bfill()

df.tail()

# Visualización del precio y del CRI
fig, ax = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

ax[0].plot(df["price"], label="Precio", color="blue")
ax[0].set_title("Serie de precios simulada")
ax[0].legend()

ax[1].plot(df["CRI"], label="CRI", color="green")
ax[1].set_title("Confidence and Risk Index (CRI)")
ax[1].set_ylim(0, 1)
ax[1].legend()

plt.tight_layout()
plt.show()

# Interpretación automática
cri_mean = df["CRI"].mean()

if cri_mean > 0.70:
    interpretation = "Régimen saludable (CRI alto)"
elif cri_mean > 0.40:
    interpretation = "Régimen de transición (CRI medio)"
else:
    interpretation = "Régimen de riesgo (CRI bajo)"

interpretation
```
