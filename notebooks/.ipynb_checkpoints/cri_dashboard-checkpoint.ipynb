# Dashboard Interactivo del Confidence and Risk Index (CRI)

Este dashboard permite:

- Visualizar el CRI por activo
- Ver regímenes y señales
- Comparar curvas de equity
- Explorar el CRI Universe de forma interactiva
- Ajustar rangos temporales y activos

```python
# Importaciones
import yfinance as yf
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display
```

```python
# Universo de activos
tickers = {
    "SPX": "^GSPC",
    "NDX": "^NDX",
    "EURUSD": "EURUSD=X",
    "BTCUSD": "BTC-USD",
    "XAUUSD": "GC=F"
}

tickers
```

```python
# Funciones del CRI
def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

def compute_cri(df):
    df = df.copy()
    df["smooth"] = smooth(df["price"])
    df["vol"] = local_volatility(df["price"])
    df["dir"] = directional_consistency(df["price"])

    vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
    dir_norm = df["dir"].rolling(10, min_periods=1).mean()

    df["CRI"] = 0.5 * (1 - vol_norm) + 0.5 * dir_norm
    df["CRI"] = df["CRI"].bfill()

    return df

def classify_regime(cri):
    if cri > 0.70:
        return "saludable"
    elif cri > 0.40:
        return "transicion"
    else:
        return "riesgo"
```

```python
# Descarga robusta
def safe_download(ticker, max_retries=5, pause=2):
    for attempt in range(max_retries):
        try:
            df = yf.download(ticker, start="2015-01-01", progress=False, auto_adjust=True)
            if not df.empty:
                return df
        except:
            pass
        time.sleep(pause)
    return None
```

```python
# Construcción del CRI Universe
data = {}

for name, ticker in tickers.items():
    df_raw = safe_download(ticker)
    if df_raw is None:
        continue

    if isinstance(df_raw.columns, pd.MultiIndex):
        df_raw = df_raw.xs(ticker, level=1, axis=1)

    df_asset = df_raw[["Close"]].rename(columns={"Close": "price"}).dropna()
    data[name] = compute_cri(df_asset)

data.keys()
```

```python
# Widgets del dashboard
selector_activo = widgets.Dropdown(
    options=list(data.keys()),
    value="SPX",
    description="Activo:"
)

selector_rango = widgets.IntSlider(
    value=365,
    min=90,
    max=2000,
    step=30,
    description="Días:"
)

display(selector_activo, selector_rango)
```

```python
# Función de actualización del dashboard
def actualizar_dashboard(activo, rango):
    df = data[activo].tail(rango)

    fig = go.Figure()

    fig.add_trace(go.Scatter(
        x=df.index,
        y=df["price"],
        name="Precio",
        line=dict(color="black")
    ))

    fig.add_trace(go.Scatter(
        x=df.index,
        y=df["CRI"],
        name="CRI",
        yaxis="y2",
        line=dict(color="green")
    ))

    fig.update_layout(
        title=f"Dashboard CRI — {activo}",
        yaxis=dict(title="Precio"),
        yaxis2=dict(title="CRI", overlaying="y", side="right", range=[0,1]),
        height=500
    )

    fig.show()

widgets.interact(actualizar_dashboard, activo=selector_activo, rango=selector_rango)
```

```python
# Heatmap del CRI Universe
cri_matrix = pd.DataFrame({
    name: df["CRI"].tail(200) for name, df in data.items()
})

fig = px.imshow(
    cri_matrix.T,
    aspect="auto",
    color_continuous_scale="RdYlGn",
    title="Heatmap CRI Universe"
)

fig.show()
```
