# # CRI Universe: visión multiactivo del Confidence and Risk Index (CRI)

Este notebook construye un “CRI Universe” multiactivo:

- Descarga de un universo amplio de activos con yfinance (MultiIndex)
- Cálculo del CRI para cada activo
- Construcción de una matriz CRI (activos × tiempo)
- Heatmap de salud del mercado
- Ranking dinámico
- Identificación de activos líderes y rezagados
- Interpretación automática por régimen

Es el panel de control global del CRI.

```python
# Importación de cotizaciones.
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Definición del universo de activos
tickers = {
    "SPX": "^GSPC",
    "NDX": "^NDX",
    "STOXX50": "^STOXX50E",
    "DAX": "^GDAXI",
    "EURUSD": "EURUSD=X",
    "GBPUSD": "GBPUSD=X",
    "USDJPY": "JPY=X",
    "BTCUSD": "BTC-USD",
    "ETHUSD": "ETH-USD",
    "XAUUSD": "GC=F",
    "XAGUSD": "SI=F",
}

tickers
```

```python
# Funciones del CRI 
def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

def compute_cri(df):
    df = df.copy()
    df["smooth"] = smooth(df["price"])
    df["vol"] = local_volatility(df["price"])
    df["dir"] = directional_consistency(df["price"])

    vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
    dir_norm = df["dir"].rolling(10, min_periods=1).mean()

    df["CRI"] = 0.5 * (1 - vol_norm) + 0.5 * dir_norm
    df["CRI"] = df["CRI"].bfill()

    return df

def classify_regime(cri):
    if cri > 0.70:
        return "Régimen saludable"
    elif cri > 0.40:
        return "Régimen de transición"
    else:
        return "Régimen de riesgo"
```

```python
# Descarga multiactivo (MultiIndex) 
raw = yf.download(
    list(tickers.values()),
    start="2015-01-01",
    progress=False,
    auto_adjust=True
)

data = {}

for name, ticker in tickers.items():
    price_series = raw["Close"][ticker].dropna()
    df_asset = price_series.to_frame(name="price")
    data[name] = df_asset

list(data.keys())
```

```python
# Cálculo del CRI para todo el universo

cri_data = {name: compute_cri(df) for name, df in data.items()}

# Comprobación rápida
cri_data["SPX"].tail()
```

```python
# Construcción de la matriz CRI Universe
# Elegimos una ventana reciente para el universo (por ejemplo, últimos 120 días)
window_days = 120

cri_universe = pd.DataFrame()

for name, df in cri_data.items():
    series = df["CRI"].tail(window_days)
    cri_universe[name] = series

cri_universe.tail()
```

```python
# Heatmap del CRI Universe
plt.figure(figsize=(12, 6))
sns.heatmap(
    cri_universe.T,
    cmap="RdYlGn",
    vmin=0,
    vmax=1,
    cbar_kws={"label": "CRI"}
)
plt.title("CRI Universe — Heatmap (últimos días)")
plt.ylabel("Activo")
plt.xlabel("Fecha")
plt.tight_layout()
plt.show()
```

```python
# Ranking actual y régimen por activo
latest_cri = cri_universe.iloc[-1]

ranking = (
    latest_cri.sort_values(ascending=False)
              .to_frame(name="CRI_ultimo")
)

ranking["Regimen"] = ranking["CRI_ultimo"].apply(classify_regime)

ranking
```

```python
# Visualización del ranking
ranking["CRI_ultimo"].plot(
    kind="barh",
    figsize=(10, 6),
    color="skyblue"
)

plt.title("Ranking CRI Universe — CRI actual por activo")
plt.xlabel("CRI")
plt.xlim(0, 1)
plt.gca().invert_yaxis()
plt.show()
```

```python
#  Identificación de líderes y rezagados
top_n = 3
bottom_n = 3

leaders = ranking.head(top_n)
laggards = ranking.tail(bottom_n)

print("Activos líderes (mayor CRI):\n")
print(leaders, "\n")

print("Activos rezagados (menor CRI):\n")
print(laggards)
```

```python
# Correlación entre CRI de activos
corr_matrix = cri_universe.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr_matrix,
    annot=False,
    cmap="coolwarm",
    center=0
)
plt.title("Correlación entre CRI de activos")
plt.tight_layout()
plt.show()
```
