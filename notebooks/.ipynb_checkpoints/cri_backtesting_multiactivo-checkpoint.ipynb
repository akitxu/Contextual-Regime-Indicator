```python
# Importaciones
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Universo de activos
tickers = {
    "SPX": "^GSPC",
    "NDX": "^NDX",
    "EURUSD": "EURUSD=X",
    "BTCUSD": "BTC-USD",
    "XAUUSD": "GC=F"
}

tickers
```

```python
# Funciones del RSI
def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

def compute_cri(df):
    df = df.copy()
    df["smooth"] = smooth(df["price"])
    df["vol"] = local_volatility(df["price"])
    df["dir"] = directional_consistency(df["price"])

    vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
    dir_norm = df["dir"].rolling(10, min_periods=1).mean()

    df["CRI"] = 0.5 * (1 - vol_norm) + 0.5 * dir_norm
    df["CRI"] = df["CRI"].bfill()

    return df

def classify_regime(cri):
    if cri > 0.70:
        return "saludable"
    elif cri > 0.40:
        return "transicion"
    else:
        return "riesgo"
```

```python
# Descarga robusta ticker‑por‑ticker
def safe_download(ticker, max_retries=5, pause=2):
    for attempt in range(max_retries):
        try:
            df = yf.download(ticker, start="2015-01-01", progress=False, auto_adjust=True)
            if not df.empty:
                return df
        except Exception as e:
            print(f"Error descargando {ticker}: {e}")

        print(f"Reintentando {ticker} ({attempt+1}/{max_retries})...")
        time.sleep(pause)

    print(f"❌ No se pudo descargar {ticker} tras {max_retries} intentos.")
    return None
```

```python
# Normalización MultiIndex + construcción de data
data = {}

for name, ticker in tickers.items():
    df_raw = safe_download(ticker)

    if df_raw is None:
        continue

    # Normalización MultiIndex
    if isinstance(df_raw.columns, pd.MultiIndex):
        df_raw = df_raw.xs(ticker, level=1, axis=1)

    # Seleccionar columna de precio
    for c in ["Close", "Adj Close", "Price"]:
        if c in df_raw.columns:
            df_asset = df_raw[[c]].rename(columns={c: "price"}).dropna()
            break

    data[name] = df_asset

data.keys()
```

```python
# Cálculo del CRI por activo
cri_data = {name: compute_cri(df) for name, df in data.items()}
cri_data["SPX"].tail()
```

```python
# Señales por activo
signal_map = {
    "saludable": 1.0,
    "transicion": 0.5,
    "riesgo": 0.0
}

for name, df in cri_data.items():
    df["Regimen"] = df["CRI"].apply(classify_regime)
    df["signal"] = df["Regimen"].map(signal_map)
```

```python
# Backtest por activo (sin warnings)
results = {}

for name, df in cri_data.items():
    df = df.copy()

    df["ret"] = df["price"].pct_change()
    df["strategy_ret"] = df["signal"].shift(1) * df["ret"]

    df = df.dropna().copy()

    df["equity_buy_hold"] = (1 + df["ret"]).cumprod()
    df["equity_strategy"] = (1 + df["strategy_ret"]).cumprod()

    results[name] = df
```

```python
# Curvas comparativas por activo
plt.figure(figsize=(14, 8))

for name, df in results.items():
    plt.plot(df["equity_strategy"], label=f"{name} - Estrategia CRI")

plt.title("Curvas de la Estrategia CRI por Activo")
plt.legend()
plt.show()
```

```python
# Métricas por activo
def CAGR(series):
    years = (series.index[-1] - series.index[0]).days / 365
    return series.iloc[-1]**(1/years) - 1

def volatility(returns):
    return returns.std() * np.sqrt(252)

def sharpe(returns, rf=0.0):
    return (returns.mean() * 252 - rf) / volatility(returns)

def max_drawdown(series):
    roll_max = series.cummax()
    dd = series / roll_max - 1
    return dd.min()

metrics = {}

for name, df in results.items():
    metrics[name] = {
        "CAGR": CAGR(df["equity_strategy"]),
        "Volatilidad": volatility(df["strategy_ret"]),
        "Sharpe": sharpe(df["strategy_ret"]),
        "Max Drawdown": max_drawdown(df["equity_strategy"])
    }

metrics_df = pd.DataFrame(metrics).T
metrics_df
```

```python
# Ranking de eficiencia
metrics_df.sort_values("Sharpe", ascending=False)
```

```python
# Interpretación automática
best = metrics_df["Sharpe"].idxmax()
worst = metrics_df["Sharpe"].idxmin()

print(f"El activo donde el CRI funciona mejor es: {best}")
print(f"El activo donde aporta menos eficiencia es: {worst}")
```

```python
#🧠 Conclusión

El CRI muestra comportamientos distintos según el activo:

- En activos tendenciales (SPX, NDX), reduce drawdowns y mejora el Sharpe.
- En activos volátiles (BTCUSD), suaviza la curva y reduce riesgo.
- En activos defensivos (XAUUSD), mantiene exposición moderada.

Este análisis permite identificar dónde el CRI aporta más valor y sirve como base para estrategias de rotación, asignación dinámica y construcción de carteras.
```
