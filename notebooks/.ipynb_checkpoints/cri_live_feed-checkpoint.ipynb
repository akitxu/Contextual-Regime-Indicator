# CRI Live Feed — Monitor en “tiempo casi real”

Este notebook:

- Descarga datos recientes de varios activos
- Calcula el CRI en cada actualización
- Muestra el régimen actual por activo
- Permite refrescar el estado bajo demanda o en bucle

```python
# Importaciones
import yfinance as yf
import numpy as np
import pandas as pd
import time
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
# Universo de activos
tickers = {
    "SPX": "^GSPC",
    "NDX": "^NDX",
    "EURUSD": "EURUSD=X",
    "BTCUSD": "BTC-USD",
    "XAUUSD": "GC=F"
}

tickers
```

```python
# Funciones CRI
def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

def compute_cri(df):
    df = df.copy()
    df["smooth"] = smooth(df["price"])
    df["vol"] = local_volatility(df["price"])
    df["dir"] = directional_consistency(df["price"])

    vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
    dir_norm = df["dir"].rolling(10, min_periods=1).mean()

    df["CRI"] = 0.5 * (1 - vol_norm) + 0.5 * dir_norm
    df["CRI"] = df["CRI"].bfill()

    return df

def classify_regime(cri):
    if cri > 0.70:
        return "saludable"
    elif cri > 0.40:
        return "transicion"
    else:
        return "riesgo"
```

```python
# Descarga rápida de ventana reciente
def download_recent(ticker, days=120):
    df = yf.download(
        ticker,
        period=f"{days}d",
        interval="1d",
        progress=False,
        auto_adjust=True
    )
    if df.empty:
        return None
    return df
```

```python
#  Cálculo CRI “live” por activo
def get_live_cri_for_universe(tickers_dict, days=120):
    data = {}
    snapshot = []

    for name, ticker in tickers_dict.items():
        df_raw = download_recent(ticker, days=days)
        if df_raw is None:
            continue

        if isinstance(df_raw.columns, pd.MultiIndex):
            df_raw = df_raw.xs(ticker, level=1, axis=1)

        if "Close" not in df_raw.columns:
            continue

        df_asset = df_raw[["Close"]].rename(columns={"Close": "price"}).dropna()
        df_cri = compute_cri(df_asset)
        data[name] = df_cri

        last_row = df_cri.iloc[-1]
        cri_val = last_row["CRI"]
        regime = classify_regime(cri_val)

        snapshot.append({
            "Activo": name,
            "Fecha": df_cri.index[-1],
            "Precio": last_row["price"],
            "CRI": cri_val,
            "Régimen": regime
        })

    snapshot_df = pd.DataFrame(snapshot).set_index("Activo")
    return data, snapshot_df
```

```python
# Widgets de control
refresh_button = widgets.Button(
    description="Refrescar ahora",
    button_style="success"
)

auto_toggle = widgets.ToggleButton(
    value=False,
    description="Auto-refresh",
    button_style="info"
)

interval_slider = widgets.IntSlider(
    value=60,
    min=10,
    max=600,
    step=10,
    description="Intervalo (s):"
)

display(refresh_button, auto_toggle, interval_slider)
```

```python
# Función de visualización
import plotly.io as pio
pio.renderers.default = "iframe"

def plot_live_cri_live(days=180):
    # Obtener datos y snapshot dentro de la función
    data, snapshot_df = get_live_cri_for_universe(tickers, days=days)

    print("Activos en data:", list(data.keys()))
    print()

    fig = go.Figure()

    for name, df in data.items():
        print(f"{name}: filas={len(df)}, columnas={df.columns.tolist()}")

        if "CRI" not in df.columns:
            print(f"  -> No hay CRI para {name}")
            continue

        if df["CRI"].dropna().empty:
            print(f"  -> CRI vacío para {name}")
            continue

        fig.add_trace(go.Scatter(
            x=df.index,
            y=df["CRI"],
            mode="lines",
            name=f"{name} - CRI"
        ))

    print("\nTrazas añadidas:", len(fig.data))

    if len(fig.data) == 0:
        print("No hay trazas válidas para graficar.")
        display(snapshot_df)
        return

    fig.update_layout(
        title="CRI por activo (ventana reciente)",
        yaxis=dict(range=[0, 1]),
        height=500
    )

    display(fig)
    display(snapshot_df.sort_index())

plot_live_cri_live(days=180)


```

```python
# Lógica de refresco manual
def on_refresh_clicked(b):
    clear_output(wait=True)
    display(refresh_button, auto_toggle, interval_slider)

    data, snapshot_df = get_live_cri_for_universe(tickers, days=180)

    print("Activos descargados:", list(data.keys()))
    print()

    plot_live_cri(data, snapshot_df)
```

```python
# Bucle de auto-refresh 
def auto_refresh_loop():
    while auto_toggle.value:
        clear_output(wait=True)
        display(refresh_button, auto_toggle, interval_slider)

        data, snapshot_df = get_live_cri_for_universe(tickers, days=180)
        plot_live_cri(data, snapshot_df)

        time.sleep(interval_slider.value)

auto_toggle.observe(lambda change: auto_refresh_loop() if change["new"] else None, names="value")
```
