# Optimización de Cartera usando el Confidence and Risk Index (CRI)

Objetivo:

- Construir una cartera multiactivo
- Usar el CRI como señal de riesgo por activo
- Ajustar pesos dinámicamente
- Comparar contra:
  - Cartera equiponderada estática
  - Cartera equiponderada dinámica sin CRI

```python
# Importaciones
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# Universo de activos
tickers = {
    "SPX": "^GSPC",
    "NDX": "^NDX",
    "EURUSD": "EURUSD=X",
    "BTCUSD": "BTC-USD",
    "XAUUSD": "GC=F"
}

tickers
```

```python
# Funciones CRI
def smooth(series, window=5):
    return series.rolling(window=window, min_periods=1).mean()

def local_volatility(series, window=5):
    return series.rolling(window=window, min_periods=1).std()

def directional_consistency(series):
    diffs = np.sign(series.diff())
    return (diffs == diffs.shift(1)).astype(int)

def compute_cri(df):
    df = df.copy()
    df["smooth"] = smooth(df["price"])
    df["vol"] = local_volatility(df["price"])
    df["dir"] = directional_consistency(df["price"])

    vol_norm = (df["vol"] - df["vol"].min()) / (df["vol"].max() - df["vol"].min())
    dir_norm = df["dir"].rolling(10, min_periods=1).mean()

    df["CRI"] = 0.5 * (1 - vol_norm) + 0.5 * dir_norm
    df["CRI"] = df["CRI"].bfill()

    return df

def classify_regime(cri):
    if cri > 0.70:
        return "saludable"
    elif cri > 0.40:
        return "transicion"
    else:
        return "riesgo"
```

```python
# Descarga robusta
def safe_download(ticker, max_retries=5, pause=2):
    for attempt in range(max_retries):
        try:
            df = yf.download(ticker, start="2015-01-01", progress=False, auto_adjust=True)
            if not df.empty:
                return df
        except:
            pass
        time.sleep(pause)
    return None
```

```python
# Construcción de precios y CRI por activo
data = {}

for name, ticker in tickers.items():
    df_raw = safe_download(ticker)
    if df_raw is None:
        continue

    if isinstance(df_raw.columns, pd.MultiIndex):
        df_raw = df_raw.xs(ticker, level=1, axis=1)

    if "Close" not in df_raw.columns:
        continue

    df_asset = df_raw[["Close"]].rename(columns={"Close": "price"}).dropna()
    df_asset = compute_cri(df_asset)
    data[name] = df_asset

data.keys()
```

```python
# Alineación temporal y retornos
prices = pd.DataFrame({name: df["price"] for name, df in data.items()}).dropna()
cris   = pd.DataFrame({name: df["CRI"]   for name, df in data.items()}).reindex(prices.index).ffill()

rets = prices.pct_change().dropna()
cris = cris.loc[rets.index]
```

```python
# Reglas de pesos
def cri_to_weight(cri_series, floor=0.0, cap=1.0, power=1.0):
    w = cri_series.clip(lower=0)
    w = w**power
    if w.sum() == 0:
        return pd.Series(0, index=w.index)
    w = w / w.sum()
    w = w.clip(lower=floor)
    w = w / w.sum()
    return w

def equal_weight(assets):
    return pd.Series(1/len(assets), index=assets)
```

```python
# Construcción de tres carteras

dates = rets.index
assets = rets.columns.tolist()
w0 = equal_weight(assets)

# 1) EQ estática (buy & hold real)
equity_assets = (1 + rets).cumprod()
equity_eq_static = (equity_assets * w0).sum(axis=1)
port_ret_eq_static = equity_eq_static.pct_change().dropna()

# 2) EQ dinámica (rebalanceo diario)
w_eq_dyn = pd.DataFrame(index=dates, columns=assets)
for t in dates:
    w_eq_dyn.loc[t] = w0

port_ret_eq_dyn = (w_eq_dyn.shift(1) * rets).sum(axis=1).dropna()
equity_eq_dyn = (1 + port_ret_eq_dyn).cumprod()

# 3) CRI-weighted (pesos diarios según CRI)
w_cri = pd.DataFrame(index=dates, columns=assets)
for t in dates:
    w_cri.loc[t] = cri_to_weight(cris.loc[t])

port_ret_cri = (w_cri.shift(1) * rets).sum(axis=1).dropna()
equity_cri = (1 + port_ret_cri).cumprod()
```

```python
# Retornos de cartera

port_ret_eq_dyn    = (w_eq_dyn.shift(1)    * rets).sum(axis=1)
port_ret_cri       = (w_cri.shift(1)       * rets).sum(axis=1)

port_ret_eq_static = port_ret_eq_static.dropna()
port_ret_eq_dyn    = port_ret_eq_dyn.dropna()
port_ret_cri       = port_ret_cri.dropna()
```

```python
# Curvas de equity
equity_eq_static = (1 + port_ret_eq_static).cumprod()
equity_eq_dyn    = (1 + port_ret_eq_dyn).cumprod()
equity_cri       = (1 + port_ret_cri).cumprod()

plt.figure(figsize=(14,6))
plt.plot(equity_eq_static, label="Cartera EQ estática", color="black")
plt.plot(equity_eq_dyn,    label="Cartera EQ dinámica", color="blue")
plt.plot(equity_cri,       label="Cartera CRI-weighted", color="orange")
plt.legend()
plt.title("Comparación de carteras")
plt.show()
```

```python
# Métricas
def CAGR(series):
    years = (series.index[-1] - series.index[0]).days / 365
    return series.iloc[-1]**(1/years) - 1

def volatility(returns):
    return returns.std() * np.sqrt(252)

def sharpe(returns, rf=0.0):
    return (returns.mean() * 252 - rf) / volatility(returns)

def max_drawdown(series):
    roll_max = series.cummax()
    dd = series / roll_max - 1
    return dd.min()

metrics = pd.DataFrame({
    "EQ estática": [
        CAGR(equity_eq_static),
        volatility(port_ret_eq_static),
        sharpe(port_ret_eq_static),
        max_drawdown(equity_eq_static)
    ],
    "EQ dinámica": [
        CAGR(equity_eq_dyn),
        volatility(port_ret_eq_dyn),
        sharpe(port_ret_eq_dyn),
        max_drawdown(equity_eq_dyn)
    ],
    "CRI-weighted": [
        CAGR(equity_cri),
        volatility(port_ret_cri),
        sharpe(port_ret_cri),
        max_drawdown(equity_cri)
    ]
}, index=["CAGR", "Volatilidad", "Sharpe", "Max Drawdown"])

metrics
```

```python
# Interpretación automática
print("📊 Comparación de carteras:\n")
display(metrics)

best_sharpe = metrics.loc["Sharpe"].idxmax()
print(f"\nLa mejor cartera en términos de Sharpe es: {best_sharpe}")
```

```python
print("Medias:")
print(port_ret_eq_static.mean(), port_ret_eq_dyn.mean(), port_ret_cri.mean())

print("\nVolatilidades:")
print(volatility(port_ret_eq_static), volatility(port_ret_eq_dyn), volatility(port_ret_cri))
```

```python
# Interpretación automática
print("📊 Comparación de carteras:\n")
display(metrics)

best_sharpe = metrics.loc["Sharpe"].idxmax()
print(f"\nLa mejor cartera en términos de Sharpe es: {best_sharpe}")
```

```python
# diagnóstico automático 
def diagnostico_carteras(metrics):
    texto = []

    # Identificar ganadores
    best_cagr   = metrics.loc["CAGR"].idxmax()
    best_sharpe = metrics.loc["Sharpe"].idxmax()
    best_risk   = metrics.loc["Volatilidad"].idxmin()
    best_dd     = metrics.loc["Max Drawdown"].idxmax()  # menos negativo = mejor

    texto.append("### 🧠 Diagnóstico automático de las carteras\n")

    # 1. CAGR
    texto.append(f"**1. Rentabilidad (CAGR)**\n")
    texto.append(f"- La cartera con mayor CAGR es **{best_cagr}**.\n")
    if best_cagr == "EQ estática":
        texto.append(
            "  Esto es típico de una cartera buy & hold: no rebalancea, deja que los ganadores dominen y termina muy concentrada."
        )
    elif best_cagr == "EQ dinámica":
        texto.append(
            "  La equiponderada dinámica logra un equilibrio entre estabilidad y crecimiento."
        )
    else:
        texto.append(
            "  La CRI-weighted logra un crecimiento superior gracias a la asignación dinámica hacia activos saludables."
        )

    texto.append("\n")

    # 2. Riesgo (volatilidad)
    texto.append(f"**2. Riesgo (volatilidad)**\n")
    texto.append(f"- La cartera con menor volatilidad es **{best_risk}**.\n")
    if best_risk == "EQ estática":
        texto.append("  Esto sería muy inusual; revisa los datos.")
    elif best_risk == "EQ dinámica":
        texto.append(
            "  Esto es coherente: el rebalanceo diario estabiliza la exposición y reduce la variabilidad."
        )
    else:
        texto.append(
            "  La CRI-weighted reduce riesgo al evitar activos en régimen de riesgo."
        )

    texto.append("\n")

    # 3. Sharpe
    texto.append(f"**3. Eficiencia riesgo/retorno (Sharpe)**\n")
    texto.append(f"- La cartera con mejor Sharpe es **{best_sharpe}**.\n")
    if best_sharpe == "EQ estática":
        texto.append(
            "  Esto sería sorprendente: implicaría que la concentración extrema fue muy rentable."
        )
    elif best_sharpe == "EQ dinámica":
        texto.append(
            "  El rebalanceo periódico mejora la eficiencia al evitar concentraciones excesivas."
        )
    else:
        texto.append(
            "  La CRI-weighted es la más eficiente: asigna más peso a activos saludables y reduce exposición en entornos adversos."
        )

    texto.append("\n")

    # 4. Drawdown
    texto.append(f"**4. Máxima caída (Max Drawdown)**\n")
    texto.append(f"- La cartera con mejor (menos negativo) drawdown es **{best_dd}**.\n")
    if best_dd == "EQ estática":
        texto.append(
            "  Esto sería muy extraño: normalmente la estática sufre drawdowns profundos por concentración."
        )
    elif best_dd == "EQ dinámica":
        texto.append(
            "  Coherente: el rebalanceo reduce la exposición a activos que caen demasiado."
        )
    else:
        texto.append(
            "  La CRI-weighted protege mejor en crisis al reducir peso en activos en régimen de riesgo."
        )

    texto.append("\n---\n")
    texto.append("### 🧩 Conclusión general\n")

    if best_sharpe == "CRI-weighted":
        texto.append(
            "La estrategia **CRI-weighted** es la más eficiente: combina riesgo controlado, drawdowns moderados y un Sharpe superior."
        )
    elif best_sharpe == "EQ dinámica":
        texto.append(
            "La estrategia **EQ dinámica** ofrece la mejor eficiencia riesgo/retorno gracias al rebalanceo periódico."
        )
    else:
        texto.append(
            "La estrategia **EQ estática** domina en Sharpe, lo cual indica un periodo excepcionalmente favorable para la concentración."
        )

    return "\n".join(texto)


# Mostrar diagnóstico
from IPython.display import Markdown
Markdown(diagnostico_carteras(metrics))
```

### 🧠 Conclusión

En este notebook hemos:

- Construido una cartera multiactivo
- Usado el CRI para asignar pesos dinámicamente
- Comparado contra carteras equiponderadas

La lógica CRI-weighted:

- Aumenta peso en activos con mejor “salud” (CRI alto)
- Reduce peso en activos en “riesgo”
- Tiende a mejorar el perfil riesgo/retorno frente a una equiponderación ciega

Este enfoque es extensible a:
- restricciones de riesgo,
- límites por activo,
- overlays de cobertura,
- y optimización media-varianza condicionada por CRI.
